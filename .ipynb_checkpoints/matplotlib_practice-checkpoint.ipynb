```python
# CELL 2
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# CELL 3
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# CELL 4
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# CELL 5
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# CELL 6
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# CELL 7
#To create a quick graph using the object-oriented interface method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# CELL 8
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# CELL 9
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# CELL 10 
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# CELL 11: SKILL DRILL (using OOI method make changes to chart)
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker = 'D', color = "green", linewidth = 2)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.legend(["Boston"])
ax.set_ylim(0,45)
ax.set_title("PyBer Fare by Month")
ax.grid()
```

```python
# CELL 23: SKILL DRILL
# Change the color of the markers to red and switch axes.
plt.scatter(y_axis, x_axis, color ='red', label = 'Chicago')
# Add a legend for the city of Chicago.
plt.legend()
# Add a title and axes labels.
plt.title("PyBer Fare by Month")
# Switch the axis so that the Fare($) data is on the x-axis.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add a limit for the x-axis data.
plt.xlim(0,50)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# CELL 24: Create a bubble chart using size parameter
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# CELL 25: Adjust size of marker
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# CELL #26: Refactor above code to use list comprehension
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# CELL 27: Create a Scatter Plot using OOI method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# CELL 28: Create a Bubble Chart using OOI method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# CELL 29: Using OOI, make changes to bubble chart above
# Create a Bubble Chart using OOI method:Switch the axis so that the Fare($) data is on the x-axis, change color of markers to 'sky blue', size of the markers to 5 times 
# each data point, color 20% transparent, black edge colors to circles, width of edge to 2
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = 'lightskyblue', alpha = 0.20, edgecolors='black', linewidths=2)
# Add a legend for the city of Boston.
ax.legend(['Boston'])
# Add a title, axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Add a limit for the x-axis data.
ax.set_xlim(0,50)
# Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
# CELL 30: Create a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# CELL 31: Make visual adjustments to pie chart above
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# adjust size of pie chart
plt.subplots(figsize=(8, 8))
# explode parameter
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create a Pie Chart using Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# SKILL DRILL: Use OOI to make changes to above pie chart
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,explode=explode,colors=colors, labels=x_axis,autopct='%1.1f%%',shadow=True,startangle=270, counterclock=True)
ax.invert_yaxis()
plt.show()
```
